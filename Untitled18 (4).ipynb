```python
import os, random
from PIL import Image
import matplotlib.pyplot as plt

base = "/kaggle/input/clean-augmented-anemia-dataset/New_Augmented_Anemia_Dataset"

def sample_and_show(site="Conjuctiva", n=6):
    site_root = os.path.join(base, site)

    available_splits = [d for d in os.listdir(site_root) if os.path.isdir(os.path.join(site_root, d))]
    if not available_splits:
        print(f"No splits found for {site}")
        return

    split = available_splits[0]
    print(f"Using split: {split}")

    sample_root = os.path.join(site_root, split)
    classes = [d for d in os.listdir(sample_root) if os.path.isdir(os.path.join(sample_root,d))]

    imgs = []
    for cls in classes:
        cls_dir = os.path.join(sample_root, cls)
        files = os.listdir(cls_dir)
        chosen = random.sample(files, min(n//len(classes)+1, len(files)))
        for f in chosen:
            imgs.append((os.path.join(cls_dir,f), cls))

    plt.figure(figsize=(12,6))
    for i,(p,cls) in enumerate(imgs[:n]):
        img = Image.open(p)
        plt.subplot(2, n//2, i+1)
        plt.imshow(img)
        plt.title(f"{site}-{split}-{cls}")
        plt.axis("off")
    plt.show()

sample_and_show("Conjuctiva", n=6)
sample_and_show("Finger_Nails", n=6)
sample_and_show("Palm", n=6)
```

```python
# ---------- CONFIG ----------
BASE_DIR = "/kaggle/input/clean-augmented-anemia-dataset/New_Augmented_Anemia_Dataset"
SITE = "Conjuctiva"   #
IMG_SIZE = (224,224)
BATCH_SIZE = 16
EPOCHS = 10
AUTOTUNE = tf.data.AUTOTUNE
MODEL_OUTDIR = "./saved_models"
os.makedirs(MODEL_OUTDIR, exist_ok=True)
```

```python
 #---------- Helpers ----------
def get_image_label_pairs(root):
    pairs = []
    classes = sorted([d for d in os.listdir(root) if os.path.isdir(os.path.join(root,d))])
    class_to_idx = {c:i for i,c in enumerate(classes)}
    for cls in classes:
        cls_dir = os.path.join(root, cls)
        for f in os.listdir(cls_dir):
            if f.lower().endswith((".jpg",".jpeg",".png")):
                pairs.append((os.path.join(cls_dir,f), class_to_idx[cls]))
    return pairs, classes

def decode_img(img_path, label):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)  # convert to RGB if grayscale
    img.set_shape([None,None,3])
    img = tf.image.resize(img, IMG_SIZE)
    img = img / 255.0
    return img, label

def augment(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, 0.1)
    img = tf.image.random_contrast(img, 0.1)
    return img, label

def build_dataset(pairs, shuffle=True, augment_flag=False):
    paths = [p for p,l in pairs]
    labels = [l for p,l in pairs]
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(paths))
    ds = ds.map(lambda p,l: decode_img(p,l), num_parallel_calls=AUTOTUNE)
    if augment_flag:
        ds = ds.map(augment, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds
```

```python
import os

def resolve_split_path(base_dir, site, split_candidates=["train","val","test","Train","Validation","Test"]):

    site_root = os.path.join(base_dir, site)
    if not os.path.exists(site_root):
        raise FileNotFoundError(f"Site folder not found: {site_root}")

    available = os.listdir(site_root)
    for cand in split_candidates:
        if cand in available:
            return os.path.join(site_root, cand)
    raise FileNotFoundError(f"No valid split folder found in {site_root}. Found: {available}")
```

```python
SITE = "Conjuctiva"

train_root = resolve_split_path(BASE_DIR, SITE, ["train","Training"])
val_root   = resolve_split_path(BASE_DIR, SITE, ["val","Validation"])
test_root  = resolve_split_path(BASE_DIR, SITE, ["test","Testing"])

train_pairs, classes = get_image_label_pairs(train_root)
val_pairs, _ = get_image_label_pairs(val_root)
test_pairs, _ = get_image_label_pairs(test_root)

print("Classes:", classes)
print("Counts:", len(train_pairs), len(val_pairs), len(test_pairs))
```

```python
import tensorflow as tf

# ---------- Augmentation ----------
def augment(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_flip_up_down(img)
    img = tf.image.random_brightness(img, max_delta=0.1)
    img = tf.image.random_contrast(img, lower=0.8, upper=1.2)
    return img, label

# ---------- Build Dataset ----------
def build_dataset(pairs, shuffle=True, augment_flag=False, batch_size=32, img_size=(224,224)):
    def _parse_function(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, img_size)
        img = img / 255.0
        return img, label

    paths = [p for p, l in pairs]
    labels = [l for p, l in pairs]

    dataset = tf.data.Dataset.from_tensor_slices((paths, labels))
    dataset = dataset.map(_parse_function, num_parallel_calls=tf.data.AUTOTUNE)

    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(paths))

    if augment_flag:
        dataset = dataset.map(augment, num_parallel_calls=tf.data.AUTOTUNE)

    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset
```

```python
import matplotlib.pyplot as plt

def preview_dataset(dataset, class_names, n=9):
    plt.figure(figsize=(10,10))
    for imgs, labels in dataset.take(1):
        for i in range(n):
            ax = plt.subplot(3, 3, i+1)
            plt.imshow(imgs[i].numpy())
            plt.title(class_names[labels[i].numpy()])
            plt.axis("off")
    plt.show()
```

```python
train_ds = build_dataset(train_pairs, shuffle=True, augment_flag=True)
val_ds   = build_dataset(val_pairs, shuffle=False, augment_flag=False)
test_ds  = build_dataset(test_pairs, shuffle=False, augment_flag=False)
```

```python
preview_dataset(train_ds, classes, n=9)
```

```python
 #---------- Build Model ----------
base = tf.keras.applications.EfficientNetB0(include_top=False, input_shape=(IMG_SIZE[0],IMG_SIZE[1],3), weights='imagenet')
base.trainable = False  # freeze first
inputs = keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
x = base(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='swish')(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation='sigmoid')(x)   # probability
model = keras.Model(inputs, outputs)
model.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss='binary_crossentropy',
              metrics=[keras.metrics.BinaryAccuracy(name='accuracy'),
                       keras.metrics.AUC(name='auc'),
                       keras.metrics.Precision(name='precision'),
                       keras.metrics.Recall(name='recall')])
model.summary()
```

```python
# callbacks
checkpoint_path = os.path.join(MODEL_OUTDIR, f"{SITE}_best.h5")
callbacks = [
    keras.callbacks.ModelCheckpoint(checkpoint_path, monitor='val_auc', mode='max', save_best_only=True, verbose=1),
    keras.callbacks.ReduceLROnPlateau(monitor='val_auc', factor=0.5, patience=3, verbose=1, mode='max'),
    keras.callbacks.EarlyStopping(monitor='val_auc', patience=6, mode='max', restore_best_weights=True, verbose=1)
]
```

```python
from sklearn.utils import class_weight
import numpy as np

y_train = [l for p, l in train_pairs]

class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))

print("Class weights:", class_weights)
```

```python
# ---------- Train ----------
history = model.fit(train_ds,
                    validation_data=val_ds,
                    epochs=10,
                    class_weight=class_weights,
                    callbacks=callbacks)

# fine-tune: unfreeze some layers
base.trainable = True
# unfreeze top N layers
fine_tune_at = len(base.layers) - 30
for layer in base.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(optimizer=keras.optimizers.Adam(1e-5),
              loss='binary_crossentropy',
              metrics=[keras.metrics.BinaryAccuracy(name='accuracy'),
                       keras.metrics.AUC(name='auc'),
                       keras.metrics.Precision(name='precision'),
                       keras.metrics.Recall(name='recall')])

history_ft = model.fit(train_ds,
                       validation_data=val_ds,
                       epochs=10,
                       class_weight=class_weights,
                       callbacks=callbacks)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

y_true = []
y_pred = []
y_prob = []

for images, labels in test_ds:
    probs = model.predict(images, verbose=0)
    preds = (probs >= 0.5).astype(int)

    y_true.extend(labels.numpy())
    y_pred.extend(preds.flatten())
    y_prob.extend(probs.flatten())

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)

# 2) Classification report
print(" Classification Report:")
print(classification_report(y_true, y_pred))

# 3) Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0","Class 1"], yticklabels=["Class 0","Class 1"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# 4) ROC Curve
fpr, tpr, _ = roc_curve(y_true, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
